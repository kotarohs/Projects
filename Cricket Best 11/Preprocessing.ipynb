```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json') as r:
    data = json.load(r)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard' : 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
match_ids_dict = {}
for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
with open('t20_json_files/t20_wc_batting_summary.json') as r:
    data = json.load(r)
    
batting = []
for rec in data:
    batting.extend(rec['battingSummary'])

df_batting = pd.DataFrame(batting)
df_batting.head()
```

```python
df_batting["out/not_out"] = df_batting["dismissal"].apply(lambda dismissal: "out" if dismissal else "not_out")
df_batting.drop(columns = ["dismissal"], inplace = True)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda batsmanName: batsmanName.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda batsmanName: batsmanName.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('t20_csv_files/t20_wc_batting_summary.csv', index = False)
```

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as r:
    data = json.load(r)
    
bowling = []
for rec in data:
    bowling.extend(rec['bowlingSummary'])

df_bowling = pd.DataFrame(bowling)
df_bowling.head()
```

```python
df_bowling["match_id"] = df_bowling["match"].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/t20_wc_bowling_summary.csv', index = False)
```
